### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda
```

#1. Análisis de facturación



```python
#Calcula el total de las ventas por tienda
tiedas = [tienda, tienda2, tienda3, tienda4]
total_ventas = []
for tienda in tiedas:
    total_ventas.append(tienda['Precio'].sum())

#Imprime el total de ventas por tienda
for i in range(len(tiedas)):
    print(f"La tienda {i+1} ha vendido {total_ventas[i]:,.2f}")
```

```python
#Mostrar la tienda con mas ingresos
print(f"La tienda {total_ventas.index(max(total_ventas))+1} ha vendido {max(total_ventas):,.2f}")
```

# 2. Ventas por categoría

```python
#Calculamos la cantidad de productos vendidos por categoria de cada tienda

count_tienda = tienda['Categoría del Producto'].value_counts()
count_tienda2 = tienda2['Categoría del Producto'].value_counts()
count_tienda3 = tienda3['Categoría del Producto'].value_counts()
count_tienda4 = tienda4['Categoría del Producto'].value_counts()

#Creamos una lista con los conteos de las categorias vendidas por tienda
list_group = [count_tienda, count_tienda2, count_tienda3, count_tienda4]
```

```python
#Imprimos la categoria mas vendida por tienda
for i in range(len(list_group)):
    print('La categoria mas vendida en la tienda ', i+1, ' es: ', list_group[i].index[0])
```

# 3. Calificación promedio de la tienda


```python
#Calculamos la calificacion promedio de de los productos por tienda
cal_tienda = tienda['Calificación'].mean()
cal_tienda2 = tienda2['Calificación'].mean()
cal_tienda3 = tienda3['Calificación'].mean()
cal_tienda4 = tienda4['Calificación'].mean()

#Creamos una lista de calificaciones promedio de cada tienda
list_cal = [cal_tienda, cal_tienda2, cal_tienda3, cal_tienda4]
```

```python
#Imprimimos la calificacion promedio con solo 2 decimales de cada tienda
for cal in range(len(list_cal)):
    print(f"La calificacion promedio de la tienda {cal+1} es: {list_cal[cal]:.2f}")

#La tienda con mayor calificacion promedio
print(f"\nLa tienda {list_cal.index(max(list_cal))+1} con mayor calificacion promedio es: {max(list_cal):.2f}")
```

# 4. Productos más y menos vendidos

```python
#Realizamos el conteo de los productos por tienda
conteo_prod_tienda = tienda['Producto'].value_counts()
conteo_prod_tienda2 = tienda2['Producto'].value_counts()
conteo_prod_tienda3 = tienda3['Producto'].value_counts()
conteo_prod_tienda4 = tienda4['Producto'].value_counts()

#Creamos una lista con los conteos de los productos por tienda
list_conteo = [conteo_prod_tienda, conteo_prod_tienda2, conteo_prod_tienda3, conteo_prod_tienda4]
```

```python
#Imprimimos los productos mas y menos vendidos por tienda

for tinda in range(len(list_conteo)):
    print(f'De la tienda {tinda+1} los productos mas vendidos son: {list_conteo[tinda].idxmax()}')
    print(f'De la tienda {tinda+1} los productos menos vendidos son: {list_conteo[tinda].idxmin()}\n')
```

# 5. Envío promedio por tienda

```python
#Realizar el calculo del costo de envio promedio por tienda
promedio_envio_tienda = round(tienda['Costo de envío'].mean(),2)
promedio_envio_tienda2 = round(tienda2['Costo de envío'].mean(),2)
promedio_envio_tienda3 = round(tienda3['Costo de envío'].mean(),2)
promedio_envio_tienda4 = round(tienda4['Costo de envío'].mean(),2)

#Creamos una lista con los promedios de costo de envio de cada tienda
promedios_envio = [promedio_envio_tienda, promedio_envio_tienda2, promedio_envio_tienda3, promedio_envio_tienda4]
```

```python
#Imprimimos los promedios de costo de envio de cada tienda
print("Promedio de costo de envío por tienda:")
for i in range(len(promedios_envio)):
    print(f"Tienda {i+1}: {promedios_envio[i]}")
```

# Graficos

```python
#Grafico de barras del total de ventas por tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold']
plt.bar(tiendas,total_ventas, color=colors)
plt.xlabel('Tienda')
plt.ylabel('Total de Ventas')
plt.title('Total de Ventas por Tienda')
ax = plt.gca()
ax.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
# Graficamos los promedios de envios como puntos conectados
plt.plot(tiendas, promedios_envio, marker='o', linestyle='-', color='purple')

plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Usamos autopct para mostrar el porcentaje de cada porción
bars = plt.barh(tiendas, list_cal, color=colors)
plt.title('Distribución Porcentual de la Calificación Promedio por Tienda')
plt.xlim(min(list_cal) * 0.98, max(list_cal) * 1.01)
plt.show()
```
